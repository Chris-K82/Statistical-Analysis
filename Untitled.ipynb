```python
import numpy as np
import statsmodels as sm
import scipy.stats as stats
import pylab
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
test = np.random.normal(20, 5, 100)
test_2 = np.random.normal(20, 5, 100)
plt.boxplot(test)
```

```python
sb.distplot(test)
```

```python
sb.kdeplot(test, kernel=['sin', 'cos', 'tri'])
```

```python
f, axes = plt.subplots(3, 4, figsize=(20, 10), sharex=True)
sb.kdeplot( test , color="skyblue", ax=axes[0, 0])
sb.distplot( test , color="olive", ax=axes[0, 1])
sb.distplot( test , color="teal", ax=axes[0, 2])
sb.distplot( test , color="grey", ax=axes[0, 3])
```

```python
sb.scatterplot(test, test_2)
```

```python
sb.regplot(test, test_2)
```

```python
stats.probplot(test, dist=stats.binom(100, 0.9), plot=pylab)
pylab.show()
```
